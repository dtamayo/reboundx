```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rcParams
import rebound
```

```python
def random_powerlaw(min_val, max_val, slope):
    """
    Generate a random number between min_val and max_val
    following a power-law distribution with given slope.
    """
    y = np.random.uniform(0., 1.)
    if slope == -1:
        return np.exp(y * np.log(max_val / min_val) + np.log(min_val))
    else:
        return (( (max_val**(slope + 1.) - min_val**(slope + 1.)) * y + min_val**(slope + 1.) ) 
                ** (1. / (slope + 1.)))
```

```python
min_val = 0.05
max_val = 0.1
slopes = [2, 3, 4, 5]
frags = np.zeros(1000)
np.random.seed(1)

for i in range(len(slopes)):
    for j in range(1000):
        frags[j] = random_powerlaw(min_val, max_val, slopes[i])
    plt.hist(frags, histtype='step', bins = 20, label = f'slope = {slopes[i]}')

plt.legend()
```
